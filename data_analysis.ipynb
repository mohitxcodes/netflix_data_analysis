```python
df["Vote_Average"] = pd.to_numeric(
    df["Vote_Average"].astype(str).str.replace(",", ""),
    errors="coerce"
).astype("float")
df["Vote_Count"] = pd.to_numeric(
    df["Vote_Count"].astype(str).str.replace(",", ""),
    errors="coerce"
).astype("Int64")


df.describe()
df.info()
# df["Vote_Count"] = df["Vote_Count"].round(0).astype("Int64")
```

```python
df.loc[
    pd.to_datetime(df["Release_Date"], errors="coerce").isna(),
    "Release_Date"
] = pd.NaT

df['Release_Date'] = pd.to_datetime(
    df["Release_Date"],
    errors="coerce"
)
```

```python
#Converting release data from object to date time and taking only year as per questions

df['Release_Date'] = pd.to_datetime(df['Release_Date'])

df['Release_Date'] = df['Release_Date'].dt.year.astype("Int64");

df['Release_Date'].dtype
```

```python
cols = ['Original_Language', 'Overview', 'Poster_Url']

df.drop(cols, axis=1, inplace=True);
```

```python
df.head()
```

```python
bins = [0,4,6,7.5,10]
labels = ["Unpopular" , "Below_Avg" , "Average" , "Popular"]

df['Vote_Average'] = pd.cut(
    df['Vote_Average'],
    bins=bins,
    labels=labels,
)
```

```python
df.head()
```

```python
df['Vote_Average'].value_counts()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.head()
```

```python
df['Genre'] = df['Genre'].str.split(', ')
df = df.explode('Genre')
```
