```python
import matplotlib.pyplot as plt;
import pandas as pd;
```

```python
df = pd.read_csv("final_moviedb.csv")
df.head()
```

```python
df['Genre'] = df['Genre'].astype('category')
```

```python
# plt.bar(df['Genre'].unique, df['Genre'].value_counts())
# plt.show()
df['Genre'].describe()
df["Genre"].value_counts()
```

```python
df['Genre'].value_counts().sort_values().plot(kind="barh", color="blue", label="Genre" )
plt.ylabel("Frequncy")
plt.title("Distrubtions of Genre")
plt.legend()
plt.tight_layout()
plt.show()
```
